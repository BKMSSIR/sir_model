```python
# import os
# print(os.getcwd())

import sys
sys.path.append('/Users/hyunjun/sir_model/')

import matplotlib.pyplot as plt

from hyunjun.sir_package import sir_function as sf
```

```python
N = 7800 * (10**6)
b0 = 0
beta = 0.7
gamma = 0.2
hs = 0.1

sus_li, inf_li, rec_li = sf.SIR(N, b0, beta, gamma, hs)

f = plt.figure(figsize=(8,5)) 
plt.plot(sus_li, 'b.', label='susceptible');
plt.plot(inf_li, 'r.', label='infected');
plt.plot(rec_li, 'c.', label='recovered/deceased');
plt.title("SIR model")
plt.xlabel("time", fontsize=10);
plt.ylabel("Fraction of population", fontsize=10);
plt.legend(loc='best')
plt.xlim(0,1000)
plt.savefig('SIR_example.png')
plt.show()


```
